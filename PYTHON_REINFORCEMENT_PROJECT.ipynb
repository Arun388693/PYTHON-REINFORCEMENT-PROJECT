## Data Cleaning

```python
# The feature - 'Unnamed: 0' is just a row index.
# So the colunm 'Unnamed: 0' is dropped.

LaptopPrice_df = laptop.drop("Unnamed: 0", axis=1)
LaptopPrice_df.head(10)
```

```python
# Check the unique values for Inches column

LaptopPrice_df['Inches'].unique()
```

```python
# Fix the unique values

LaptopPrice_df['Inches']=LaptopPrice_df['Inches'].str.replace('?', '',regex = False)
```

```python
LaptopPrice_df['Inches']=pd.to_numeric(LaptopPrice_df['Inches'], errors='coerce')
```

```python
# Fix the unique values for Ram column

LaptopPrice_df['Ram']=LaptopPrice_df['Ram'].str.replace('GB', '')
```

```python
LaptopPrice_df['Ram']=pd.to_numeric(LaptopPrice_df['Ram'], errors='coerce')
```

```python
# Fix the unique values for memory column

LaptopPrice_df['Memory']=LaptopPrice_df['Memory'].str.replace('?','', regex=False)
```

```python
# Fix the unique values for weight column

LaptopPrice_df['Weight']=LaptopPrice_df['Weight'].str.replace('kg', '')
```

```python
LaptopPrice_df['Weight']=pd.to_numeric(LaptopPrice_df['Weight'], errors='coerce')
LaptopPrice_df
```

```python
# Check null values

LaptopPrice_df.isnull().sum()
```

```python
LaptopPrice_df.isnull().sum().sum()
```

```python
# Numeric columns - fill with median
LaptopPrice_df['Inches']=LaptopPrice_df['Inches'].fillna(LaptopPrice_df['Inches'].median())
LaptopPrice_df['Ram']=LaptopPrice_df['Ram'].fillna(LaptopPrice_df['Ram'].median())
LaptopPrice_df['Weight']=LaptopPrice_df['Weight'].fillna(LaptopPrice_df['Weight'].median())
LaptopPrice_df['Price']=LaptopPrice_df['Price'].fillna(LaptopPrice_df['Price'].median())
```

```python
# Text/Mixed columns â†’ fill with mode
LaptopPrice_df['Cpu']=LaptopPrice_df['Cpu'].fillna(LaptopPrice_df['Cpu'].mode()[0])
LaptopPrice_df['ScreenResolution']=LaptopPrice_df['ScreenResolution'].fillna(LaptopPrice_df['ScreenResolution'].mode()[0])
LaptopPrice_df['Memory']=LaptopPrice_df['Memory'].fillna(LaptopPrice_df['Memory'].mode()[0])
LaptopPrice_df['Company']=LaptopPrice_df['Company'].fillna(LaptopPrice_df['Company'].mode()[0])
LaptopPrice_df['TypeName']=LaptopPrice_df['TypeName'].fillna(LaptopPrice_df['TypeName'].mode()[0])
LaptopPrice_df['Gpu']=LaptopPrice_df['Gpu'].fillna(LaptopPrice_df['Gpu'].mode()[0])
LaptopPrice_df['OpSys']=LaptopPrice_df['OpSys'].fillna(LaptopPrice_df['OpSys'].mode()[0])
```

```python
LaptopPrice_df.isnull().sum()
```

```python
LaptopPrice_df.shape
```

```python
# Find Duplicates
LaptopPrice_df.duplicated().sum()
```

```python
LaptopPrice_df=LaptopPrice_df.drop_duplicates()
```

```python
LaptopPrice_df.duplicated().sum()
```

```python
LaptopPrice_df
```

```python
# Deal with ScreenResolution column we should extract Touchscreen and Ips and XxY resolution
LaptopPrice_df["Touchscreen"] = LaptopPrice_df["ScreenResolution"].apply(lambda x: 1 if "Touchscreen" in x else 0)
LaptopPrice_df["Ips"] = LaptopPrice_df["ScreenResolution"].apply(lambda x: 1 if "IPS" in x else 0)
LaptopPrice_df[["X_resolution", "Y_resolution"]] = LaptopPrice_df["ScreenResolution"].str.extract(r'(\d+)x(\d+)').astype(int)
LaptopPrice_df.drop(columns=["ScreenResolution"], inplace=True)
LaptopPrice_df.head()
```

```python
LaptopPrice_df[['CPU', 'Frequency']] = LaptopPrice_df['Cpu'].str.rsplit(' ', n=1, expand= True)
```

```python
LaptopPrice_df[['Storage', 'StorageCapacity']] = LaptopPrice_df['Memory'].str.rsplit(' ', n=1, expand= True)
```

```python
LaptopPrice_df
```

```python
LaptopPrice_df['Frequency']=LaptopPrice_df['Frequency'].str.replace('GHz', ''). astype(float)
LaptopPrice_df
```

```python
# Take the most common companies and label the rest as 'Other'
valid_companies = ["Apple", "Dell", "HP", "Lenovo", "Asus", "Acer", "MSI", "Toshiba"]
LaptopPrice_df["Company"] = LaptopPrice_df["Company"].where(LaptopPrice_df["Company"].isin(valid_companies), "Other")
LaptopPrice_df["Company"].value_counts()
```

## Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(x=LaptopPrice_df['Inches'])
plt.title("Boxplot")
plt.show()
```

```python
LaptopPrice_df['Inches'].describe()
```

```python
# Fix the outliers

Q1 = LaptopPrice_df['Inches'].quantile(0.25)
Q3 = LaptopPrice_df['Inches'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

mean_value = LaptopPrice_df['Inches'].mean()

LaptopPrice_df['Inches'] = np.where((LaptopPrice_df['Inches'] < lower_bound) | (LaptopPrice_df['Inches'] > upper_bound),mean_value,LaptopPrice_df['Inches'])
print(LaptopPrice_df['Inches'])
```
